## <a name="preparation"></a> Preprocessing

Séparation des données
- variable cibles
- données d'entrainement et test

Variables qualitative => dichotomisation

Variables quantitative => centrage et réduction



```python
# dichotomisation des variables qualitatives sur le jeu de données entier
list_quali  = ['At2supAt1','IT', 'Ft', 'Fm', 'Cr', 'Mp']
list_quanti = ['m', 'Mt', 'At1', 'W', 'ec', 'ep', 'Enedc', 'Ewltp']

# choix de la modalité de référence
cars_dummies = pd.get_dummies(cars[list_quali], columns = list_quali).drop(['At2supAt1_0',
                                                                            'IT_0',
                                                                            'Ft_petrol',
                                                                            'Fm_mono_fuel',
                                                                            'Cr_M1',
                                                                            'Mp_BMW GROUP'],axis=1)
cars_lm = cars[list_quanti]
cars_lm = pd.concat([cars[list_quanti], cars_dummies], axis=1)
cars_lm.columns
```

```python
# Séparation des données
target_nedc = cars['Enedc']
feats_nedc = cars_lm.drop(['Mt', 'Enedc', 'Ewltp'], axis=1)
X_nedc_train, X_nedc_test, y_nedc_train, y_nedc_test = train_test_split(feats_nedc, target_nedc,
                                                                        test_size=0.25,
                                                                        random_state=33)
target_wltp = cars['Ewltp']
feats_wltp = cars_lm.drop(['m', 'Enedc', 'Ewltp'], axis=1)
X_wltp_train, X_wltp_test, y_wltp_train, y_wltp_test = train_test_split(feats_wltp, target_wltp,
                                                                        test_size=0.25,
                                                                        random_state=33)
```

```python
# normalisation des données
sc_nedc = preprocessing.StandardScaler()
X_nedc_scaled_train = pd.DataFrame(sc_nedc.fit_transform(X_nedc_train),
                                   columns = X_nedc_train.columns,
                                   index = X_nedc_train.index)
X_nedc_scaled_test = pd.DataFrame(sc_nedc.transform(X_nedc_test),
                                   columns = X_nedc_test.columns,
                                   index = X_nedc_test.index)

sc_wltp = preprocessing.StandardScaler()
X_wltp_scaled_train = pd.DataFrame(sc_wltp.fit_transform(X_wltp_train),
                                   columns = X_wltp_train.columns,
                                   index = X_wltp_train.index)
X_wltp_scaled_test = pd.DataFrame(sc_wltp.transform(X_wltp_test),
                                  columns = X_wltp_test.columns,
                                  index = X_wltp_test.index)
```

## <a name="wltpXGBoost"></a> Modèle XGboot
### Tuning 


```python
## Fixer learning_rate et n_estimators

def modelfit(alg, dtrain, y_train, dtest, y_test, useTrainCV=True, cv_folds=10, early_stopping_rounds=50):
    
    # déterminer le nombre d'arbres    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain   = xgb.DMatrix(dtrain.values, label=y_train.values)
        xgtest    = xgb.DMatrix(dtest.values,  label=y_test.values)
        cvresult  = xgb.cv(xgb_param, xgtrain,
                           num_boost_round=alg.get_params()['n_estimators'],
                           nfold=cv_folds,
                           metrics='rmse',
                           early_stopping_rounds=early_stopping_rounds)
        display(cvresult.shape[0])
        display(cvresult.tail())
        
        plt.plot(cvresult.iloc[:,0])
        plt.ylabel("RMSE")
        plt.xlabel('Round')
        plt.show();
        
        alg.set_params(n_estimators=cvresult.shape[0])
        
    # entrainer l'algorithme sur les données
    alg.fit(dtrain, y_train)
        
    # prédictions sur les données d'entrainement
    dtrain_predictions = alg.predict(dtrain)
    
    # prédisction sur les données test
    dtest_predictions = alg.predict(dtest)
    
    # métriques
    print("\nDonnées d'entrainement")
    print("RMSE : %.4g" % np.sqrt(mean_squared_error(y_train.values, dtrain_predictions)))
    print("MAE : %.4g" % mean_absolute_error(y_train.values, dtrain_predictions))
    print("Coefficient de détermination : %.4g" % r2_score(y_train.values, dtrain_predictions))
      
    print("\nDonnées de test")
    print("RMSE : %.4g" % np.sqrt(mean_squared_error(y_test.values, dtest_predictions)))
    print("MAE : %.4g" % mean_absolute_error(y_test.values, dtest_predictions))
    print("Coefficient de détermination : %.4g" % r2_score(y_test.values, dtest_predictions))


xgb1 = XGBRegressor(
    objective='reg:squarederror',
    eval_metric='rmse',
    learning_rate=0.1,
    random_state=33)

modelfit(xgb1, X_wltp_scaled_train, y_wltp_train, X_wltp_scaled_test, y_wltp_test)
```

```python
# Tuning pour max_depth et min_child_weight
param_test1 = {
    'max_depth':range(2,16,2),
    'min_child_weight':range(1,18,2)
}
xgb2 = XGBRegressor(
    objective       = 'reg:squarederror',
    eval_metric     = 'rmse',
    learning_rate   = 0.1,
    n_estimators    = 100,
    random_state    = 33)

gsearch1 = GridSearchCV(estimator = xgb2,
                        param_grid = param_test1,
                        scoring='neg_root_mean_squared_error',
                        n_jobs=-1,
                        cv=10)

gsearch1.fit(X_wltp_scaled_train, y_wltp_train)

gsearch1_result=pd.concat([pd.DataFrame(gsearch1.cv_results_["params"]),
           pd.DataFrame(gsearch1.cv_results_["mean_test_score"],
                        columns=["neg_RMSE - mean"])]
          ,axis=1)
print ("meilleur paramètres:", gsearch1.best_params_)
print("meilleur score:", gsearch1.best_score_)
```

##  <a name="nedc"></a> Norme NEDC
### <a name="nedclineaire"></a>Modèle de régression linéaire simple puis régularisé

```python
## données sans les 3 points influents
x = X_nedc_scaled_train.drop(X_nedc_scaled_train.loc[ind].index, axis=0)
y = y_nedc_train.drop(y_nedc_train.loc[ind].index, axis=0)


print('Norme NEDC - régression linéaire non régularisée')

lr_nedc = LinearRegression()
lr_nedc.fit(x, y)
print('  - Coefficient de détermination du modèle :', lr_nedc.score(X_nedc_scaled_train, y_nedc_train).round(3))
print('  - Coefficient de détermination des données test :', lr_nedc.score(X_nedc_scaled_test,y_nedc_test).round(3))

lr_nedc_pred_train = lr_nedc.predict(X_nedc_scaled_train)
lr_nedc_pred_test = lr_nedc.predict(X_nedc_scaled_test)
print('  - RMSE train :', np.sqrt(mean_squared_error(lr_nedc_pred_train, y_nedc_train)).round(3))
print('  - RMSE test :', np.sqrt(mean_squared_error(lr_nedc_pred_test, y_nedc_test)).round(3))
print('  - MAE train :', mean_absolute_error(lr_nedc_pred_train, y_nedc_train).round(3))
print('  - MAE test :', mean_absolute_error(lr_nedc_pred_test, y_nedc_test).round(3))
```

```python
## Norme NEDC - Recherche des paramètres les plus explicatifs avec une régression ElasticNet

model_en = ElasticNetCV(cv=10, 
                        l1_ratio = (0.1, 0.25, 0.5, 0.75, 0.99),
                        n_alphas = 200,
                        random_state=33)

model_en.fit(x, y)
print('Alpha sélectionné par validation croisée', model_en.alpha_.round(4))
print('l1_ratio_: %f' % model_en.l1_ratio_)

print('\nNorme WLTP - régression linéaire régularisée')
pred_train = model_en.predict(x)
pred_test = model_en.predict(X_nedc_scaled_test)

print('  - RMSE train :', np.sqrt(mean_squared_error(y, pred_train)).round(3))
print('  - RMSE test :', np.sqrt(mean_squared_error(y_nedc_test, pred_test)).round(3))
print('  - MAE train :', mean_absolute_error(pred_train, y).round(3))
print('  - MAE test :', mean_absolute_error(pred_test, y_nedc_test).round(3))
      
print('  - Coefficient de détermination du modèle :', model_en.score(x, y).round(4))
print('  - Coefficient de détermination des données test :', model_en.score(X_nedc_scaled_test, y_nedc_test).round(4))
```
